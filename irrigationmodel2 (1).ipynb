```python
import numpy as np
import matplotlib.pyplot as plt

# Define the moisture content grid (10x10)
moisture_grid = np.random.randint(0, 100, size=(10, 10))  # Randomly generated moisture content grid

# Define the amount of water to distribute (in liters)
predicted_water = 100  # Example: 100 liters

# Function to distribute water optimally across the grid
def distribute_water(moisture_grid, predicted_water):
    # Calculate total moisture content in the grid
    total_moisture = np.sum(moisture_grid)

    # Calculate the proportion of water to allocate to each cell based on moisture content
    water_proportions = (moisture_grid / total_moisture) * predicted_water

    # Ensure that no cell receives more water than needed to reach 100% moisture content
    water_proportions = np.minimum(water_proportions, 100 - moisture_grid)

    return water_proportions

# Optimize water distribution
water_distribution = distribute_water(moisture_grid, predicted_water)

# Calculate total water distributed
total_water_distributed = np.sum(water_distribution)

# Plot the moisture content grid
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.imshow(moisture_grid, cmap='YlGnBu', interpolation='nearest')
plt.colorbar(label='Moisture Content')
plt.title('Moisture Content Grid')

# Plot the optimized water distribution
plt.subplot(1, 2, 2)
plt.imshow(water_distribution, cmap='Blues', interpolation='nearest')
plt.colorbar(label='Water Distribution (liters)')
plt.title('Optimized Water Distribution')

# Add annotations for total water distributed
for i in range(10):
    for j in range(10):
        plt.text(j, i, f'{water_distribution[i, j]:.1f}', ha='center', va='center', color='black')

# Add annotation for total water distributed
plt.text(0, -1, f'Total Water Distributed: {total_water_distributed:.1f} liters', fontsize=12, fontweight='bold')

plt.tight_layout()
plt.show()

# Plot total water distribution across rows and columns
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.bar(np.arange(10), np.sum(water_distribution, axis=1))
plt.xlabel('Row')
plt.ylabel('Total Water Distributed (liters)')
plt.title('Total Water Distribution Across Rows')

plt.subplot(1, 2, 2)
plt.bar(np.arange(10), np.sum(water_distribution, axis=0))
plt.xlabel('Column')
plt.ylabel('Total Water Distributed (liters)')
plt.title('Total Water Distribution Across Columns')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the moisture content grid (10x10)
moisture_grid = np.random.randint(0, 100, size=(10, 10))  # Randomly generated moisture content grid

# Define the amount of water to distribute (in liters)
predicted_water = 100  # Example: 100 liters

# Function to distribute water optimally across the grid
def distribute_water_optimized(moisture_grid, predicted_water):
    # Calculate the deficit moisture for each cell
    deficit_moisture = 100 - moisture_grid

    # Calculate the total deficit moisture
    total_deficit = np.sum(deficit_moisture)

    # Calculate the proportion of water to allocate to each cell based on deficit moisture
    water_proportions = (deficit_moisture / total_deficit) * predicted_water

    # Ensure that no cell receives more water than needed to reach 100% moisture content
    water_proportions = np.minimum(water_proportions, deficit_moisture)

    return water_proportions

# Optimize water distribution
water_distribution_optimized = distribute_water_optimized(moisture_grid, predicted_water)

# Calculate total water distributed
total_water_distributed_optimized = np.sum(water_distribution_optimized)

# Plot the moisture content grid
plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
plt.imshow(moisture_grid, cmap='YlGnBu', interpolation='nearest')
plt.colorbar(label='Moisture Content')
plt.title('Moisture Content Grid')

# Plot the optimized water distribution
plt.subplot(1, 2, 2)
plt.imshow(water_distribution_optimized, cmap='Blues', interpolation='nearest')
plt.colorbar(label='Water Distribution (liters)')
plt.title('Optimized Water Distribution')

# Add annotations for total water distributed
for i in range(10):
    for j in range(10):
        plt.text(j, i, f'{water_distribution_optimized[i, j]:.1f}', ha='center', va='center', color='black')

# Add annotation for total water distributed
plt.text(0, -1, f'Total Water Distributed (Optimized): {total_water_distributed_optimized:.1f} liters', fontsize=12, fontweight='bold')

plt.tight_layout()
plt.show()

# Plot total water distribution across rows and columns
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.bar(np.arange(10), np.sum(water_distribution_optimized, axis=1))
plt.xlabel('Row')
plt.ylabel('Total Water Distributed (liters)')
plt.title('Total Water Distribution Across Rows (Optimized)')

plt.subplot(1, 2, 2)
plt.bar(np.arange(10), np.sum(water_distribution_optimized, axis=0))
plt.xlabel('Column')
plt.ylabel('Total Water Distributed (liters)')
plt.title('Total Water Distribution Across Columns (Optimized)')

plt.tight_layout()
plt.show()
```
